```python
import matplotlib.pyplot as plt
import pandas as pd
file = 'https://raw.githubusercontent.com/dknife/ML/main/data/vehicle_prod.csv'

df = pd.read_csv(file, index_col = 0)
plt.figure(figsize = (12, 4))

plt.subplot (2, 2, 1)
df['2009'].plot(kind= 'line')
plt.grid()

plt.subplot (2, 2, 2)
df['2009'].plot(kind= 'pie')

plt.subplot (2, 2, 3)
df['2009'].plot(kind= 'bar', color = ('orange', 'r', 'b', 'm', 'c', 'k'))

plt.subplot (2, 2, 4)
df['2009'].plot(kind= 'area')
```
