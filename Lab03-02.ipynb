```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

filename = 'http://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data'
column_names = ['MPG', 'Cylinders', 'Displacement', 'Horsepower', 'Weight', 'Acceleration', 'Model Year', 'Origin']
mpg_df = pd.read_csv(filename, names = column_names, na_values = '?', comment = '\t', sep = ' ', skipinitialspace = True)

mpg_df.head()
```

```python
# 처음 3개의 행들을 출력하라

mpg_df.head(3)
```

```python
# 누락된 항목을 찾아서 출력하기

missing_data = mpg_df[mpg_df['Horsepower'].isna()]
missing_data
```

```python
# 누락된 항목을 가지고 있는 열을 삭제하기

pd.DataFrame.dropna(mpg_df, how = 'any',inplace = True)
mpg_df.count()
```

```python
# 누락된 항목을 그 열의 데이터들의 평균으로 대체하기

mean_horsepower = mpg_df['Horsepower'].mean()
mpg_df['Horsepower'].fillna(mean_horsepower, inplace =True)
print(mpg_df['Horsepower'].isna().sum())

print(mpg_df['Horsepower'].head(33))
```
