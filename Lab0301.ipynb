```python
import matplotlib.pyplot as plt
import numpy as np

image = plt.imread('drive/MyDrive/AI/lenna.png')
x = np.linspace(0, 10, 100)
y_sin = np.sin(0.5 * np.pi * x)
y_cos = np.cos(0.5 * np.pi * x)

plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.imshow(image)
plt.axis('off')

plt.subplot(1, 2, 2)
plt.plot(x, y_sin, 'ro--', label='sine')
plt.plot(x, y_cos, 'b^--', label='cosine')
plt.title('Graph: sin(0.5πx) and cos(0.5πx)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
