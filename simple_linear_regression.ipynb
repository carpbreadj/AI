```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

def h(x, w, b):
  return w * x + b

data_home = 'https://github.com/dknife/ML/raw/main/data/'
lin_data = pd.read_csv(data_home + 'pollution.csv')
print(lin_data.head(3), end = '\n\n')
print(lin_data.describe(), end = '\n\n')
lin_data.plot(kind = 'scatter', x = 'input', y = 'pollution')
```

```python
x = lin_data['input'].to_numpy() #numpy배열 형태로 변환
y = lin_data['pollution'].to_numpy()

w = [1, -3, -4.5445]
b = [1, 6, 5.1755]

for n in range(3):
  o = h(x, w[n], b[n])
  print(n, "-th Hypothesis Loss(MSE): {0:.4f}".format(0.5 * np.average(np.square(y - o))))
```

```python
x0 = lin_data['input'].min()
x1 = lin_data['input'].max()

lin_data.plot(kind = 'scatter', x = 'input', y = 'pollution')
#x(x축 시작과 끝, y축 시작과 끝)
plt.plot([x0, x1], [h(x0, w[0], b[0]), h(x1, w[0], b[0])], 'bo--', label = 'Hypothesis 0')
plt.plot([x0, x1], [h(x0, w[1], b[1]), h(x1, w[1], b[1])], 'go--', label = 'Hypothesis 1')
plt.plot([x0, x1], [h(x0, w[2], b[2]), h(x1, w[2], b[2])], 'ro--', label = 'Normal Eqn')

plt.grid()
plt.legend()
plt.show()
```
