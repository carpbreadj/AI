```python
import matplotlib.pyplot as plt
import numpy as np

image = plt.imread('drive/MyDrive/AI/lenna.png')
x = np.linspace(-2.5, 2.5, 20)
sigmoid = 1 / (1 + np.exp(-x))
relu = np.maximum(0, x)

plt.subplot(2, 2, 1)
plt.plot(x, sigmoid, 'rs-')
plt.title('Sigmoid')
plt.grid()

plt.subplot(2, 2, 2)
plt.plot(x, relu, 'bo-')
plt.title('Relu')
plt.grid()

plt.subplot(2, 2, 3)
plt.plot(x, sigmoid, 'r-', label = 'sigmoid')
plt.plot(x, relu, 'b--', label = 'relu')
plt.title('Sigmoid & Relu')
plt.legend()
plt.grid()

plt.subplot(2, 2, 4)
plt.imshow(image)
plt.axis('off')
plt.title('Lenna')

plt.tight_layout()

plt.show()
```
