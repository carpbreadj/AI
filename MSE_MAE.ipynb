```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

def h(x, w, b):
  return w * x + b
x = np.arange(6)
y = np.array([0, 2, 3, 8, 8, 10])

#model : f(x) = w*x, note: b = 0
mse_w = []
mae_w = []
w = np.arange(0, 4.1, 0.2)

for wc in w:
  o = h(x, wc, 0)
  mse_w.append(np.average(np.square(y - o)))
  mae_w.append(np.average(np.abs(y - o)))

plt.plot(w, mse_w, 'ro--', label = 'MSE')
plt.plot(w, mae_w, 'bo--', label = 'MAE')
plt.xlabel('w')
plt.ylabel('Loss(w)')
plt.legend()
plt.grid()
plt.show()
```
