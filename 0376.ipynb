```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 3 * np.pi, 0.2)
ysin = np.sin(0.5 * np.pi * x)
ycos = np.cos(0.5 * np.pi * x)

plt.plot(x, ysin, 'ro-', label = 'sine')
plt.plot(x, ycos, 'b^-', label = 'cosine')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph, $sin(0.5  \pi x)$ and $cos (0.5 \pi x)$')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.subplot(2, 1, 1)
plt.plot(x, y_sin)
plt.title('Sine')

plt.subplot(2, 1, 2)
plt.plot(x, y_cos)
plt.title('Cosine')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

N = 10
# (평균이 0인 정규분포, 표준편차가 1인 정규분포, 정규분포를 따르는 난수를 N개 설정)
x = np.random.normal(0, 1, N)
y = np.random.normal(5, 2, N)
y2 = np.random.normal(5, 2, N)
# (생성된 난수의 최소값 1, 생성된 난수의 최대값(포함X) 500, 생성할 난수의 개수 N)
area = np.random.randint(1, 500, N)
c = np.sqrt(area)

plt.figure(figsize = [16, 4])
plt.subplot (1, 3, 1)
plt.scatter(x, y, c = 'r')
plt.scatter(x, y2, s = 100, c = 'b') #s = size = 100으로 고정
plt.grid()

plt.subplot(1, 3, 2)
plt.scatter(x, y, s = area, c = c) #c의 값에 따라 색상 정의
plt.grid()

plt.subplot (1, 3, 3)
plt.scatter(x, y, s = area, c = c, marker = '^')
plt.grid()
```
